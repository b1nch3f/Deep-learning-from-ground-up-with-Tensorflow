### Linear Models May Go Wrong

- To find whether an image is Cat or Dog, we can't rely on the intensity of individual pixels. We've to know the neighbouring pixel values.

### From Linear to Nonlinear

$
\begin{split}
    \begin{aligned}
        \mathbf{H} & = \sigma(\mathbf{X} \mathbf{W}^{(1)} + \mathbf{b}^{(1)}), \\
        \mathbf{O} & = \mathbf{H}\mathbf{W}^{(2)} + \mathbf{b}^{(2)}
    \end{aligned}
\end{split}
$

### Universal approximators

- We can approximate any functions $Y = f(X)$ using deeper (vs. wider) networks.

### Activation Functions

#### ReLU Function

$\operatorname{ReLU}(x) = \max(x, 0).$

```python
%matplotlib inline
import tensorflow as tf
from d2l import tensorflow as d2l
```

```python
x = tf.Variable(tf.range(-8.0, 8.0, 0.1), dtype=tf.float32)
y = tf.nn.relu(x)
d2l.plot(x.numpy(), y.numpy(), 'x', 'relu(x)', figsize=(5, 2.5))
```

```python
with tf.GradientTape() as t:
    y = tf.nn.relu(x)
d2l.plot(x.numpy(), t.gradient(y, x).numpy(), 'x', 'grad of relu',
         figsize=(5, 2.5))
```

#### parameterized ReLU (pReLU)

$\operatorname{pReLU}(x) = \max(0, x) + \alpha \min(0, x)$

#### Sigmoid Function

$\operatorname{sigmoid}(x) = \frac{1}{1 + \exp(-x)}$

```python
y = tf.nn.sigmoid(x)
d2l.plot(x.numpy(), y.numpy(), 'x', 'sigmoid(x)', figsize=(5, 2.5))
```

##### derivative of the sigmoid function

$\frac{d}{dx} \operatorname{sigmoid}(x) = \frac{\exp(-x)}{(1 + \exp(-x))^2} = \operatorname{sigmoid}(x)\left(1-\operatorname{sigmoid}(x)\right)$

```python
with tf.GradientTape() as t:
    y = tf.nn.sigmoid(x)
d2l.plot(x.numpy(), t.gradient(y, x).numpy(), 'x', 'grad of sigmoid',
         figsize=(5, 2.5))
```

#### Tanh Function

$\operatorname{tanh}(x) = \frac{1 - \exp(-2x)}{1 + \exp(-2x)}$

```python
y = tf.nn.tanh(x)
d2l.plot(x.numpy(), y.numpy(), 'x', 'tanh(x)', figsize=(5, 2.5))
```

##### derivative of the tanh function

$\frac{d}{dx} \operatorname{tanh}(x) = 1 - \operatorname{tanh}^2(x)$

```python
with tf.GradientTape() as t:
    y = tf.nn.tanh(x)
d2l.plot(x.numpy(), t.gradient(y, x).numpy(), 'x', 'grad of tanh',
         figsize=(5, 2.5))
```
